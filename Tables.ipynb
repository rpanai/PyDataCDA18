```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.DataFrame({"csv":[142, 753 ],
                   "parq":[106, 560],
                   "parqSRTD":[40, 197]}, index=["10M", "50M"])
```

```python
df.plot(kind="bar", rot=0,
        title="Size in Disk").legend(loc='center left', bbox_to_anchor=(1, 0.89));
```

```python
df = pd.DataFrame({"csv":[3.94, 34],
                   "csvSRTD":[3.02, 15.2],
                   "parq":[2.99, 18.5],
                   "parqSRTD":[2.59, 13.3]}, index=["10M", "50M"])
```
